```python
import numpy as nop
import pandas as pd
import cv2
from PIL import Image
import subprocess
import os
```

```python
data_path = "/scratch/ab9738/pollution_img/mirflickr1m/"
```

```python
exif_path = data_path + "exif/"
```

```python
image_path = data_path + "images/"
```

```python
folders = [str(i) for i in range(10,20)]
```

```python
folders
```

```python
all_exif_files = []
all_image_files = []
for folder in folders:
    exif_folder_path = exif_path + folder + "/"
    image_folder_path = image_path + folder + "/"
    exif_file_list = os.listdir(exif_folder_path)
    exif_file_path_list = [exif_folder_path + i for i in exif_file_list]
    all_exif_files += exif_file_path_list
    
    image_file_list = os.listdir(image_folder_path)
    image_file_path_list = [image_folder_path + i for i in image_file_list]
    all_image_files += image_file_path_list
```

```python
exif_files_with_time = []
```

```python
# for file in all_exif_files:
def check_time(file):
    try:
        with open(file) as f:
            data = f.read()
        f.close()
        if("-Date and Time (Original)" in data):
            return True
    except:
        return False
    return False
```

```python
from joblib import Parallel, delayed
```

```python
exif_time_bool = Parallel(n_jobs=12)(delayed(check_time)(file) for file in all_exif_files)
```

```python
for i, flag in enumerate(exif_time_bool):
    if(flag):
        exif_files_with_time += [all_exif_files[i]]
```

```python
len(exif_files_with_time)
```

```python
label_dict = {}
```

```python
for i, file in enumerate(exif_files_with_time):
    f = open(file)
    data = f.read()
    data = data.split('\n')
    index = data.index("-Date and Time (Original)")
    timestr = data[index+1].split(" ")[-1][:8]
    try:
        time_in_sec = int(timestr.split(":")[0])*3600 + int(timestr.split(":")[1])*60 + int(timestr.split(":")[2])
        time_100 = time_in_sec/864.0
        tmp_file = file.replace("exif", "images")
        img_file = tmp_file.replace("txt", "jpg")
        label_dict[img_file] = time_100
    except:
        continue
```

```python
len(label_dict)
```
