```python
import pandas as pd

insta = pd.read_csv('../insta_data.csv')
insta['dt_round'] = pd.to_datetime(insta['dt_round'])
insta['datetime'] = pd.to_datetime(insta['datetime'])
insta['dt_round_2'] = insta['datetime'].dt.round('60min')
insta.head()
```

```python
insta['dt_round_2'] = insta['dt_round_2'].dt.tz_convert(None)
insta['dt_round_2'] = insta['dt_round_2'].dt.tz_localize('Asia/Kolkata')
insta.head()
```

```python
poll = pd.read_csv('../combined_del_pol.csv')
poll['From Date'] = pd.to_datetime(poll['From Date'],dayfirst=True)
poll['To Date'] = pd.to_datetime(poll['To Date'],dayfirst=True)
poll.head()
```

```python
poll['From Date'] = poll['From Date'].dt.tz_localize('Asia/Kolkata')
poll['To Date'] = poll['To Date'].dt.tz_localize('Asia/Kolkata')
poll.head()
```

```python
insta['ppm'] = None
insta.head()
```

```python
# delhi junction railway station = mandir marg, sonia vihar
# Indira Gandhi International Airport, indira gandhi airport = airport
# humayun's tomb, Humayun's tomb = lodhi road/ major dhyan chand
# india gate, India Gate, Rajpath, India Gate, New Delhi, Rajpath, Raisina Hill, rajpath India Gate, = major dhyan chand
# old delhi = ito delhi
# gurgaon/gurgram, Cyber Hub Gurgaon = aya nagar, delhi
# DLF One Horizon Center = aya nagar, delhi
# Nehru stadium = lodhi road
# Okhla Industrial Area III, okhla bird sanctuary = CRRI mathura road
# Cirrus9, Rooftop Bar At The Oberoi, Delhi = lodhi road/JLN stadium
# Chawri Bajar Delhi - 110006 = ito delhi
# chandni chowk = ito delhi
# yamuna river = ito delhi
# noida = sector-62 noida + sector 125-noida
# pritampura = wazirpur
# jama masjid = ito delhi
# defence colony = nehru nagar, jln
# nehu place, nehru place = nehru nagar/crri mathura
# saket = sri aurobindo marg
# lodhi garden = lodhi road
# rajpath = major dhyan chand
# Purana Qila = major dhyan chand
# DND flyway = sector 125 - noida
# connaught place = mandir marg
# others = avg(from above)

def avg_pm(row_list):
    sum_lst = 0
    count = 0
    for i in row_list[2:]:
        if (i is not None) and (i != "") and (i != "None"):
            sum_lst += float(i)
            count += 1
    if count == 0:
        return None
    return sum_lst/count

def two_pm(val1, val2):
    pm = None
    if (val1 is not None) and (val1 != "") and (val1 != "None"):
        if (val2 is not None) and (val2 != "") and (val2 != "None"):
            pm = (float(val1)+float(val2))/2
        else:
            pm = float(val1)
    else:
        if (val2 is not None) and (val2 != "") and (val2 != "None"):
            pm = float(val2)
    return pm

def one_pm(val):
    pm = None
    if (val is not None) and (val != "") and (val != "None"):
        pm = float(val)
    return pm

drop_idx = []
for index, row in insta.iterrows():
    loc = str(row['location']).lower()
    time = row['dt_round_2']
    x = poll.loc[poll['From Date'] == time]
    pm = None
    if x.empty:
        drop_idx.append(index)
        continue
    elif loc == "delhi junction railway station":
        pm = two_pm(x.iloc[0]["Mandir Marg"],x.iloc[0]["Sonia Vihar"])
    elif "airport" in loc:
        pm = one_pm(x.iloc[0]["IGI Airport (T3)"])
    elif "humayun" in loc:
        pm = two_pm(x.iloc[0]["Major Dhyan Chand National Stadium"],x.iloc[0]["Lodhi Road"])
    elif ("india gate" in loc) or ("rajpath" in loc) or ("qila" in loc):
        pm = one_pm(x.iloc[0]["Major Dhyan Chand National Stadium"])
    elif ("gurgaon" in loc) or ("gurugram" in loc) or ("horizon center" in loc) or ("cyberhub" in loc):
        pm = one_pm(x.iloc[0]["Aya Nagar"])
    elif ("nehru stadium" == loc) or ("jln stadium" in loc) or ("lodhi" in loc):
        pm = one_pm(x.iloc[0]["Lodhi Road"])
    elif "okhla" in loc:
        pm = one_pm(x.iloc[0]["CRRI Mathura Road"])
    elif "cirrus9" in loc:
        pm = two_pm(x.iloc[0]["Jawaharlal Nehru Stadium"],x.iloc[0]["Lodhi Road"])
    elif (loc == "old delhi") or ("chawri" in loc) or ("chandni" in loc) or ("yamuna" in loc) or ("jama masjid" in loc):
        pm = one_pm(x.iloc[0]["ITO"])
    elif "noida" in loc:
        pm = two_pm(x.iloc[0]["Sector - 125"],x.iloc[0]["Sector - 62"])
    elif "pritampura" in loc:
        pm = one_pm(x.iloc[0]["Wazirpur"])
    elif "defence colony" in loc:
        pm = two_pm(x.iloc[0]["Nehru Nagar"],x.iloc[0]["Jawaharlal Nehru Stadium"])
    elif ("nehu place" in loc) or ("nehru place" in loc):
        pm = two_pm(x.iloc[0]["Nehru Nagar"],x.iloc[0]["CRRI Mathura Road"])
    elif "saket" in loc:
        pm = one_pm(x.iloc[0]["Sri Aurobindo Marg"])
    elif "dnd" in loc:
        pm = one_pm(x.iloc[0]["Sector - 125"])
    elif "connaught" in loc:
        pm = one_pm(x.iloc[0]["Mandir Marg"])
    if pm is None:
        pm = avg_pm(list(x.iloc[0]))
    insta.set_value(index,'ppm',pm)

print(drop_idx)
insta.head()
```

```python
insta = insta.drop(drop_idx)
len(insta)
```

```python
def classify(x): #x is float
    if x < 12:
        return 0
    elif x < 35.5:
        return 1
    elif x < 55.5:
        return 2
    elif x < 150.5:
        return 3
    elif x < 250.5:
        return 4
    elif x < 350.5:
        return 5
    elif x < 1000:
        return 6
    return -1

insta["class"] = -1
for index, row in insta.iterrows():
    insta.set_value(index,'class',classify(float(row["ppm"])))
insta
```
