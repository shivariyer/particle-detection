```python
import pandas as pd

del1 = pd.read_csv('../delhi_pol1.csv')
print(len(del1))
del1.head()
```

```python
del2 = pd.read_csv('../delhi_pol2.csv')
print(len(del2))
del2.head()
```

```python
del3 = pd.read_csv('../delhi_pol3.csv')
print(len(del3))
del3.head()
```

```python
for area in list(del2.columns.values)[2:]:
    del1[area] = del2[area]

for area in list(del3.columns.values)[2:]:
    del1[area] = del3[area]

print(len(del1))
del1.head()
```

```python
del1.to_csv('../combined_del_pol.csv',index=False)
```
