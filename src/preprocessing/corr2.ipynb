```python
from scipy.stats import spearmanr
from scipy.stats import pearsonr

def corr(data):
    per50 = []
    per75 = []
    per80 = []
    per90 = []
    permean = []
    permax = []
    ppm = []
    for index, row in data.iterrows():
        filename = '../img/'+row['filename']+'_trans.jpg' #_mod_trans.jpg
        #city = row['filename'][0]
        #if city == "B":
        trans = cv2.imread(filename,0)
        ppm.append(row['ppm'])
        per50.append(np.percentile(trans,50))
        per75.append(np.percentile(trans,75))
        per80.append(np.percentile(trans,80))
        per90.append(np.percentile(trans,90))
        permean.append(np.average(trans))
        permax.append(np.max(trans))
           
    print("corr median")
    print(pearsonr(ppm,per50))
    print(spearmanr(ppm,per50))
    print("corr 75th percentile")
    print(pearsonr(ppm,per75))
    print(spearmanr(ppm,per75))
    print("corr 80th percentile")
    print(pearsonr(ppm,per80))
    print(spearmanr(ppm,per80))
    print("corr 90th percentile")
    print(pearsonr(ppm,per90))
    print(spearmanr(ppm,per90))
    print("corr avg")
    print(pearsonr(ppm,permean))
    print(spearmanr(ppm,permean))
    print("corr max")
    print(pearsonr(ppm,permax))
    print(spearmanr(ppm,permax))
```

```python
# Delhi
corr(delhi)
```
