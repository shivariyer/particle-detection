```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [20, 10]

img = cv2.imread('../img/42118043_297985804327105_62667303486013762_nframe9.jpg',1)
dst1 = cv2.fastNlMeansDenoisingColored(img,None,40,40,7,21)

plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(dst1)
```

```python
dst2 = cv2.fastNlMeansDenoisingColored(img,None,10,10,7,21)

plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(dst2)
```

```python
blur = cv2.GaussianBlur(img,(5,5),0)

plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(blur)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
plt.subplot(1, 2, 1), plt.imshow(gray)
plt.subplot(1, 2, 2), plt.imshow(thresh)
```
